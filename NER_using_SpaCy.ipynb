**Named Entity Recognition using spacy**

```python
import spacy
nlp=spacy.load('en_core_web_sm')
import string
```

There are three pipelines in spacy namely tagger, parser, NER

```python
nlp.pipe_names
```

```python
text1=(""" Mahabharata and Ramayana are the epics  of India based on true stories of ancient times. Mahabharata tells us about the very famous battle of Kurukshetra .The
 Bhagwat Gita is part of this great epic .Arjuna and Krishna are two among
 the many heroes of this epic. Ramayana contains the reason behind celebrating festivals like Dusshera and Diwali. Ramayana talks about Rama ,Sita ,Ravana ,Hanuman and their battle of Lanka.
""").lower()
```

```python
doc1=nlp(text1)
for ent in doc1.ents:
  print(ent.text,ent.start_char,ent.end_char,ent.label_)
```
